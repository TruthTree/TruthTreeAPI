```python
import pandas as pd
```

```python
csv_data = pd.read_csv(
    './assets/StateData.csv',
    error_bad_lines=False)
df = pd.DataFrame(csv_data)
```

```python
df.iloc[0]['SurveyYr']
```

```python
columns = ['Name', 'FIPS_Code_State']
df_state = pd.DataFrame(columns=columns)
```

```python
for i in range(51):
    current = df.iloc[i]
    df_state.at[i, 'Name'] = current['Name']
    df_state.at[i, 'FIPS_Code_State'] = current['FIPS_Code_State']
```

```python
state_population_columns = ['State_id', 'Population', 'Year']
df_state_population = pd.DataFrame(columns=state_population_columns)
```

```python
for year in range(2016, 2006, -1):
    for state in df_state['Name'].unique():
        df_current = df[(df['Year4']==year) & (df['Name']==state)]
        df_to_append = pd.DataFrame(columns=state_population_columns)
        df_to_append['Year'] = df_current['Year4'].fillna(0)
        df_to_append['State_id'] = df_state[(df['Name']==state)].index.item()
        df_to_append['Population'] = df_current['Population'].fillna(0)
        df_state_population = df_state_population.append(df_to_append, ignore_index=True)
        
        
```
